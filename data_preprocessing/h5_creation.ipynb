```python
import os
import pickle
import h5py
import numpy as np
from glob import glob
```

```python
pickle_dir = '/home/mdelafuente/SSL/out_directory/'
```

```python
pickle_files = glob(os.path.join(pickle_dir, "*.parquet"))
    
if not pickle_files:
    raise ValueError(f"No pickle files found in {pickle_dir}")

# First, collect all unique keys and their array shapes
print("Analyzing pickle files...")
all_keys = set()
sample_arrays = {}
```

```python
# Load first file to get keys and array shapes
with open(pickle_files[0], 'rb') as f:
    first_dict = pickle.load(f)
    all_keys.update(first_dict.keys())
    sample_arrays = {k: v for k, v in first_dict.items()}
```
